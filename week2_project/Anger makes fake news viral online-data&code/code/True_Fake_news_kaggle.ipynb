```python
import pandas as pd
import numpy as np
import re
import spacy


fake_news_title = pd.read_csv('data/Kaggle/fake_news_emo_title.csv')
fake_news_1 = pd.read_csv('data/Kaggle/fake_news_emo_1.csv')
fake_news_2 = pd.read_csv('data/Kaggle/fake_news_emo_2.csv')
fake_news_3 = pd.read_csv('data/Kaggle/fake_news_emo_3.csv')
fake_news = pd.concat([fake_news_1, fake_news_2, fake_news_3])
fake_news = fake_news.reset_index(drop=True)
fake_news['emo_words_title'] = fake_news_title['emo_words_title']
true_news = pd.read_csv('data/Kaggle/true_news_emo.csv')
```

```python
import ast
true_news['emo_words_title'] = true_news['emo_words_title'].apply(lambda d: ast.literal_eval(d))
true_news['emo_words_text'] = true_news['emo_words_text'].apply(lambda d: ast.literal_eval(d))
fake_news['emo_words_title'] = fake_news['emo_words_title'].apply(lambda d: ast.literal_eval(d))
fake_news['emo_words_text'] = fake_news['emo_words_text'].apply(lambda d: ast.literal_eval(d))
```

```python
emos = ['anger', 'anticipation', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'trust', 'positive', 'negative']

for emo in emos:
    true_news[emo] = true_news.apply(lambda ew: ew['emo_words_title'][emo]+ew['emo_words_text'][emo], axis=1)
    fake_news[emo] = fake_news.apply(lambda ew: ew['emo_words_title'][emo]+ew['emo_words_text'][emo], axis=1)
```

```python
special_emos = ['anger', 'anticipation', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'trust']
true_news['label'] = 'true'
fake_news['label'] = 'fake'
# fake_news = fake_news.drop('index',axis=1)
news = true_news.append(fake_news)
news['special_emos_sum'] = news[special_emos].sum(axis=1)
news.head()
```

```python
special_emos_dist = ['anger_d', 'anticipation_d', 'disgust_d', 'fear_d', 'joy_d', 'sadness_d', 'surprise_d', 'trust_d']
news = news[news['special_emos_sum']!=0]
news[special_emos_dist] = news[special_emos]
trans = news[special_emos_dist].T/news[special_emos_dist].T.sum()
news[special_emos_dist] = trans.T
```

```python
news['anger_joy_d'] = news['anger_d'] - news['joy_d']
```
