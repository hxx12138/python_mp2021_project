```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data01 = pd.read_excel('data/0-1.xls')
data02 = pd.read_excel('data/0-2.xls')
data11 = pd.read_excel('data/1-1.xls')
data12 = pd.read_excel('data/1-2.xls')
data11.head()
```

```python
l = [data01, data02, data11, data12]
l[0].head()
```

```python
for i in range(len(l)):
    l[i]['SUM1'] = l[i].loc[:,'M1-1':'M1-6'].sum(axis=1)
    l[i]['SUM2'] = l[i].loc[:,'M2-1':'M2-6'].sum(axis=1)
    l[i]['SUM3'] = l[i].loc[:,'M3-1':'M3-5'].sum(axis=1)
    l[i]['SUM4'] = l[i].loc[:,'M4-1':'M4-4'].sum(axis=1)
data01.head()
```

```python
data01[['SUM1','SUM2','SUM3','SUM4']].to_csv('data/data01.csv', index=False)
data02[['SUM1','SUM2','SUM3','SUM4']].to_csv('data/data02.csv', index=False)
data11[['SUM1','SUM2','SUM3','SUM4']].to_csv('data/data11.csv', index=False)
data12[['SUM1','SUM2','SUM3','SUM4']].to_csv('data/data12.csv', index=False)
```
